```python
import pandas as pd
import dateutil.parser as parser
from ntscraper import Nitter 
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
scraper = Nitter()
```

```python
tweets = scraper.get_tweets('elonmusk' , mode ='user' ,number = 50)
```

```python
final_tweets=[]
for tweet in tweets['tweets']:
    data=[tweet['link'] ,tweet['text'] ,tweet['date'] , tweet['stats']['likes'] , tweet['stats']['comments'] ]
    final_tweets.append(data)
    
```

```python
final_tweets
```

```python
df = pd.DataFrame(final_tweets ,columns =['link' , 'text' , 'date' ,'No_of_likes' ,'No_of_tweets'])
```

```python
df
```

#### Removing Unnecessary Column 
###### (ColumnThat Does Not Add Value)

```python
columns_to_remove = ['link']
df.drop(columns=columns_to_remove, inplace=True)
```

```python
df
```

#### Extracting additional features from the ‚Äòtext‚Äô column 

Word Count:

```python
df['word_count'] = df['text'].apply(lambda x: len(x.split()))
```

```python
df
```

insert a new column that calculates the average number of likes per tweets 

```python
Average_No_of_likes = df ['No_of_likes'] / df['No_of_tweets']
df.insert(5 ,'Average_No_of_likes',Average_No_of_likes )
```

```python
df
```

Calculate the total likes and tweets for each month

```python
df['date'] = pd.to_datetime(df['date'], format='%b %d, %Y ¬∑ %I:%M %p UTC')
### converting the date column --> to datetime 
```

```python
df
```

```python
df_month = df.groupby(pd.Grouper(key='date', freq='M'))
df_month_sum = df_month.agg({'No_of_likes': 'sum', 'No_of_tweets': 'sum'})

# Print the results
print(df_month_sum)
```

```python
df['day_of_week'] = df['date'].dt.day_name()
```

```python
df
```

```python
df['day_of_week'] = df['date'].dt.weekday
```

```python
df
```

```python
low_threshold = 1000
high_threshold = 60000

# Create a new column 'like_category'
df['like_category'] = pd.cut(df['No_of_tweets'], bins=[0, low_threshold, high_threshold, float('inf')],
                             labels=['low', 'medium', 'high'])
```
